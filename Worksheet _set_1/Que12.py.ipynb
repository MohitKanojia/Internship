12. Write a python program to find whether a number is prime or composite

```python
num=int(input("Enter the no:"))
count=0
i=1
while i<=num:
    if num%i==0:
        count=count+1
        i=i+1
        
    if count==2:
         print("Its a Prime No:")
    elif count>2:
         print("Its a Composite No:")
    else:
         print("The Number is neither prime or composite")
```
